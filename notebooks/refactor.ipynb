```python
import django
import os

os.environ['DJANGO_SETTINGS_MODULE'] = 'dashboard.settings'
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
django.setup()
```

```python
from precinct.models import Precinct, PrecinctMap
```

```python
for p in Precinct.objects.all():
    pm = PrecinctMap.objects.get(county__county_code=p.county.county_code, precinct_short_name=p.precinct_id, edition=p.edition)
    pm.precinct = p
    pm.save()

```
